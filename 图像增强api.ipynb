```python
#resize
#crop
#flip
#brightness & contrast
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior() 
```

```python
name='./yaoshuige.jpg'
img_string=tf.read_file(name)
img_decoded=tf.image.decode_image(img_string)
sess=tf.Session()
img_decoded_val=sess.run(img_decoded)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
#resize 

#tf.image.resize_area
#tf.image.resize_bicubic
#tf.image.resize_nearest_neighbor
name='./yaoshuige.jpg'
img_string=tf.read_file(name)
img_decoded=tf.image.decode_image(img_string)
img_decoded=tf.reshape(img_decoded,[1,313,500,3])

resize_img=tf.image.resize_bicubic(img_decoded,[626,1000])

sess=tf.Session()
img_decoded_val=sess.run(resize_img)
img_decoded_val=img_decoded_val.reshape((626,1000,3))
img_decoded_val=np.asarray(img_decoded_val,np.uint8)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
#crop
#tf.image.pad_to_bounding_box
#tf.image.crop_to_bounding_box
#tf.random_crop
name='./yaoshuige.jpg'
img_string=tf.read_file(name)
img_decoded=tf.image.decode_image(img_string)
img_decoded=tf.reshape(img_decoded,[1,313,500,3])

padded_img=tf.image.pad_to_bounding_box(img_decoded,50,100,500,800)

sess=tf.Session()
img_decoded_val=sess.run(padded_img)
img_decoded_val=img_decoded_val.reshape((500,800,3))
img_decoded_val=np.asarray(img_decoded_val,np.uint8)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
#flip
#tf.image.flip_up_down
#tf.image.flip_left_right
#tf.image.random_flip_up_down
#tf.image.random_flip_left_right
name='./yaoshuige.jpg'
img_string=tf.read_file(name)
img_decoded=tf.image.decode_image(img_string)
img_decoded=tf.reshape(img_decoded,[1,313,500,3])

flipped_img=tf.image.flip_up_down(img_decoded)

sess=tf.Session()
img_decoded_val=sess.run(flipped_img)
img_decoded_val=img_decoded_val.reshape((313,500,3))
img_decoded_val=np.asarray(img_decoded_val,np.uint8)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
#brightness
#tf.image.adjust_brightness
#tf.image.random_brightness
#tf.image.adjust_constrast
#tf.image.random_constrast
name='./yaoshuige.jpg'
img_string=tf.read_file(name)
img_decoded=tf.image.decode_image(img_string)
img_decoded=tf.reshape(img_decoded,[1,313,500,3])

new_img=tf.image.adjust_brightness(img_decoded,0.5)

sess=tf.Session()
img_decoded_val=sess.run(new_img)
img_decoded_val=img_decoded_val.reshape((313,500,3))
img_decoded_val=np.asarray(img_decoded_val,np.uint8)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```
